1. Explain Relative, Absolute and Mixed Cell Referencing


   -Relative Reference (A1):
     In this type, the cell reference changes when we copy the formula to another cell.
     Example: =A1+B1 becomes =A2+B2 when copied down


   -Absolute Reference ($A$1):
     In this type, the cell reference does not change when copied.
     Example: =$A$1+$B$1  


   -Mixed Reference (A$1 or $A1):
     In this type, one part is fixed and the other part changes.
     Example: A$1 means row is fixed but column can change.
